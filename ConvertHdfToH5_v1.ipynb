- The goal is to convert CALIPSO files in hdf format into an hdf5 format that can be open by python

```python
import os
import numpy as np
```

```python
#/Volumes/DAGORETBACK/CALIPSO/DATA/ICARE/CALIOP/L3/CAL_LID_L3_APro_CloudySkyTransparent.v3.00/CAL_LID_L3_APro_CloudySkyTransparent.v3.00
top_path_root='/Volumes/DAGORETBACK/CALIPSO/DATA/ICARE/CALIOP/L3/CAL_LID_L3_APro_CloudFree.v3.00/CAL_LID_L3_APro_CloudFree.v3.00'
```

```python
years=np.arange(2007,2016,1)
```

```python
olddir= os.getcwd()
```

```python
for year in years:
    fullpath_root=os.path.join(top_path_root,str(year))
    print fullpath_root
    
    os.chdir(fullpath_root)
    # Check current working directory.
    retval = os.getcwd()
    print "Current working directory %s" % retval
    
    for fn in os.listdir('.'):
        command='h4toh5 '+fn
        print (command)
        os.system(command)
    
    if not os.path.exists('hdf5'):
        os.makedirs('hdf5')
    os.system('mv *.h5 hdf5/.')

    if not os.path.exists('hdf4'):
        os.makedirs('hdf4')
    os.system('mv *.hdf hdf4/.')
    
```
